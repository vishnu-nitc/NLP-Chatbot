```python
from text_preprocessing import preprocess_text
from text_preprocessing import to_lower, remove_stopword, remove_punctuation, lemmatize_word,remove_special_character,keep_alpha_numeric
```

```python
df['Description']=df['Description'].apply(lambda x: remove_stopword(x))
df['Description']=df['Description'].apply(lambda x: lemmatize_word(x))
```

```python
df['Description']=df['Description'].apply(lambda x: [remove_punctuation(i) for i in x])
df['Description']=df['Description'].apply(lambda x: [remove_special_character(i) for i in x])
```

```python
print('Average length of Description : ',df.Description.str.len().mean())
print('Maximum length of Description : ',df.Description.str.len().max())
print('Minimum length of Description : ',df.Description.str.len().min())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Employee or Third Party']=le.fit_transform(df['Employee or Third Party'])
df['Critical Risk']=le.fit_transform(df['Critical Risk'])
df['Genre']=le.fit_transform(df['Genre'])
df['Industry Sector']=le.fit_transform(df['Industry Sector'])
```

```python
df.drop(columns=df.columns[0],axis=1,inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Data'])

df['Year'] = df.Date.apply(lambda x : x.year)
df['Month'] = df.Date.apply(lambda x : x.month)
df['Day'] = df.Date.apply(lambda x : x.day)
df['Weekday'] = df.Date.apply(lambda x : x.day_name())
df['WeekofYear'] = df.Date.apply(lambda x : x.weekofyear)
```

```python
#Defining seasons for months in Brazil
def month2seasons(x):
    if x in [9, 10, 11]:
        season = 'Spring'
    elif x in [12, 1, 2]:
        season = 'Summer'
    elif x in [3, 4, 5]:
        season = 'Autumn'
    elif x in [6, 7, 8]:
        season = 'Winter'
    return season
```

```python
df['Season'] = df['Month'].apply(month2seasons)
```

```python
df['Season']=le.fit_transform(df['Season'])
```

```python
from tqdm import tqdm
embeddings_index = {}
EMBEDDING_FILE = '/content/drive/My Drive/AIML/glove.6B.300d.txt'
f = open(EMBEDDING_FILE)
for line in tqdm(f):
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```
